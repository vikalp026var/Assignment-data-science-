```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x=tips.tip,y=tips.total_bill)
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.jointplot(x=tips.tip,y=tips.total_bill)
```

```python
sns.pairplot(tips)
```

```python
sns.relplot(x=tips.tip,y=tips.total_bill,data=tips,style='size')
```
